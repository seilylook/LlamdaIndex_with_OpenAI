# Data Chunk

```python
def get_chunk(text, chunk_size):
    words = text.split()
    chunks = [" ".join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]

    return chunks
```

```python
chunks = get_chunk(text, 128)
chunks
```

# Text Embeding

```python
def get_embedding(text):
    return client.embeddings.create(input=[text], model="text-embedding-3-small").data[0].embedding
    
```

```python
embeddings = [get_embedding(c) for c in chunks]
embeddings
```

# Retrieve

```python
pip install numpy
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def retriever(query, top_k):
    q_emb = get_embedding(query)

    similar_score = cosine_similarity([q_emb], embeddings)[0]
    max_indices = np.argsort(similar_score)[::-1][:top_k]

    retrieved_datas = [chunks[i] for i in max_indices]

    return retrieved_datas
```

```python
contexts = retriever("OpenAI의 sora 모델에 대해 설명해줘", 3)
contexts
```

# DocumentSummaryIndex

문서를 요약해서 저장하기 때문에 LLM이 필요하다.

```python
from llama_index.core import DocumentSummaryIndex

chatgpt = OpenAI(model="gpt-3.5-turbo")

doc_summary_index = DocumentSummaryIndex.from_documents(
    documents,
    llm=chatgpt,
    show_progress=True
)
```

```python
from llama_index.core.indices.document_summary import (
    DocumentSummaryIndexLLMRetriever,
)
```

# Generator

```python
from llama_index.core import get_response_synthesizer

response_synthesizer = get_response_synthesizer(response_mode="compact")

response = response_synthesizer.synthesize(
    "Mistral AI는 어느 국가에 속해 있어?", nodes=retriever.retrieve("Mistral AI는 어느 국가에 속해 있어?")
)

response.response
```

response_synthesizer를 정의할 때는 여러가지 모드를 설정할 수 있다.

가장 기본으로 사용되는 모드는 compact이다.

compact 모드는 chunk들을 몇개씩 결합해서 Prompt에 보낸다.

다음 Chunk들을 얻으면 새로 얻은 정보로 이전의 대답을 Refine 하게 된다.

이전의 대답인 '이거'고 새로운 대답은 '이거'인데, 두 가지를 참고했을 때 대답을 뭘까하는 방식이다.

Refine도 비슷한 모드인데, 여러 Chunk들을 결합하는 대신에 Chunk 하나당 한번에 LLM 콜을 하게 된다.

```python
response_synthesizer = get_response_synthesizer(response_mode="accumulate")

response = response_synthesizer.synthesize(
    "Mistarl AI는 어느 국가에 속해 있어?", nodes=retriever.retrieve("Mistarl AI는 어느 국가에 속해 있어?")
)
response.response
```

accumulate 모드에서는 각각의 chunk 마다 서로 다른 대답을 생성한다.

각각의 chunk마다 다른 대답을 볼 수 있다.

```python
response_synthesizer = get_response_synthesizer(response_mode="simple_summarize")

response = response_synthesizer.synthesize(
    "Mistarl AI는 어느 국가에 속해 있어?", nodes=retriever.retrieve("Mistarl AI는 어느 국가에 속해 있어?")
)
response.response
```

simple_summarize는 받은 chunk들을 한 번에 LLM 콜의 context에 들어갈 수 있도록 truncate한다.

chunk 사이즈가 큰 데 중요한 정보가 chunk의 뒤쪽에 들어가 있으면 상당한 정보 손실이 있을 수 있다.

```python
response_synthesizer = get_response_synthesizer(response_mode="tree_summarize")

response = response_synthesizer.synthesize(
    "Mistarl AI는 어느 국가에 속해 있어?", nodes=retriever.retrieve("Mistarl AI는 어느 국가에 속해 있어?")
)
response.response
```

tree_summarize 모드는 tree 방식으로 해서 노드 하나가 남을 때까지 summarize를 반복한다.

chunk 몇개를 결합해서 여기에서 쿼리에 대한 답을 내놓으라고 하고 그 답들을 chunk로 삼아서 

새로운 chunke들을 주고 여기에서 쿼리에 대한 답을 내놓으라고 하는 방식이다.

```python
from llama_index.core import PromptTemplate


new_prompt_qa = """
Context information is below.
---------------------
{context_str}
---------------------
Given the context information and not prior knowledge, answer the query.
Use proper Korean, and answer in the style of kindergarten teacher, gentle and enthusiastic.
Query: {query_str}
Answer: 
"""

new_prompt_refine = """
The original query is as follows: {query_str}
We have provided an existing answer: {existing_answer}
We have the opportunity to refine the existing answer (only if needed) with some more context below.
------------
{context_msg}
------------
Given the new context, refine the original answer to better answer the query. If the context isn't useful, return the original answer.
Use proper Korean, and answer in the style of kindergarten teacher, gentle and enthusiastic.
Refined Answer: 
"""

new_prompt_qa = PromptTemplate(new_prompt_qa)
new_prompt_refine = PromptTemplate(new_prompt_refine)


query_engine.update_prompts(
    {"response_synthesizer:text_qa_template": new_prompt_qa, "response_synthesizer:refine_template": new_prompt_refine}

)
```
